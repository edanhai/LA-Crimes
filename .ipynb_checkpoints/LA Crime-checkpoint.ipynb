# Load Data

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
crime = pd.read_csv('Crime_Data_from_2020_to_Present_20250108.csv')
```

```python
df = crime.drop(labels=['Mocodes','Weapon Used Cd','Weapon Desc','Crm Cd 2','Crm Cd 3','Crm Cd 4','Cross Street'],axis=1)
```

```python
df
```

```python
df.sort_values(by='Date Rptd')
```

```python
df.dtypes
```

```python
df['Date Rptd']=pd.to_datetime(df['Date Rptd'],format='%Y-%m-%d')	
df['DATE OCC']= pd.to_datetime(df['DATE OCC'],format='%Y-%m-%d')
df['TIME OCC']= pd.to_datetime(df['TIME OCC'],format='%H%M%S',errors='coerce')
```

```python
df
```

```python
df.info()
```

```python
df.describe(include=['float64','int64'])
```

```python
df['Vict Age'].value_counts()
```

```python
df_age = df[df['Vict Age'] > 0]
```

```python
df_age['Vict Age'].value_counts()
```

```python
df['year'] = df['DATE OCC'].dt.year
df['month'] = df['DATE OCC'].dt.month
```

```python
df
```

```python
df.sort_values(by='Date Rptd')
```

# Feature Understanding

```python
df['month'].value_counts().sort_index().plot()
```

```python
df.groupby(['DATE OCC','Crm Cd Desc']).size()
```

```python
crime_coun_per_day  = df.groupby(['DATE OCC','Crm Cd Desc']).size()
```

```python
crime_coun_per_day.groupby('Crm Cd Desc').mean().reset_index(name='average_daily_count')
```

```python
top_20_crimes = crime_coun_per_day.groupby('Crm Cd Desc').mean()\
                    .reset_index(name='average_daily_count').sort_values(by='average_daily_count',ascending=False).head(20)
```

```python
top_20_crimes
```

```python
plt.barh(top_20_crimes['Crm Cd Desc'],top_20_crimes['average_daily_count'])
```

```python
top_20_crimes_sorted = top_20_crimes.sort_values(by='average_daily_count',ascending=True)
```

```python
plt.barh(top_20_crimes_sorted['Crm Cd Desc'],top_20_crimes_sorted['average_daily_count'])
```

```python
plt.style.available
```

```python
plt.style.use('ggplot')
```

```python
plt.figure(figsize=(12,8))
#line chart
plt.subplot(2,2,1)
crime_plot = df.groupby(['year','month']).size().plot(color='green',linewidth=2)
crime_plot.set_xlabel('Year- Month')
crime_plot.set_title('Crime over time')

#hist 
plt.subplot(2,2,2)
plt.hist(df_age['Vict Age'],edgecolor='black',bins=15)
plt.tight_layout()
plt.title('Distribution of Victim Age')

#horizontal
plt.figure()
plt.barh(top_20_crimes_sorted['Crm Cd Desc'],top_20_crimes_sorted['average_daily_count'],color='orange')
plt.title('Top 20 Crimes')
```
